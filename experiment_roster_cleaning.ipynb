```python
#Cleaning dataset Experiment ROster
#Que informacion tiene? Tiene solo dos columnas: client_id y variation, que te indica que cliente esta en el grupo control o en el grupo test. 
```

```python
#1. Importar dataset

import pandas as pd
df_exp = pd.read_csv("Datasets/df_final_experiment_clients.txt")
```

```python
#2.a. Exploracion del dataset
df_exp.info()
```

```python
df_exp.shape
```

```python
df_exp.head()
```

```python
#2.b Revisar los valores nulos
df_exp.isnull().sum()
#Importante: en la columna Variation tenemos un total de 20109 valores nulos. 
#Aqui podemos considerar varias hipotesis: error de datos, clientes no asignados o clientes no incluidos en el experimento. 
```

```python
#Vamos a comprobar los valores que presentan la variable Variation. 
df_exp["Variation"].value_counts(dropna=False)
```

```python
#Aqui podemos ver que los 20109 valores nulos, no son errores. Sino que, en el experimento no participaron todos los clientes activos de Vanguard, sino una muestra fue asignada al proyecto. 
#Con este resultado, considero que la decision seria eliminar los valores NaN del analisis, porque no aporta informacion al objetivo del analisis: analizar el A/B testing entre grupo test y grupo control. 
```

```python
#Vamos a eliminar los valores Nan de la columna Variation.
df_exp = df_exp.dropna(subset=["Variation"])
```

```python
#Comprobacion de que se han borrado correctamente. 
df_exp["Variation"].value_counts(dropna=False)
```

```python
#2.c Revisar los duplicados.
#En este caso, vamos a revisar que la columna de client_id no presenta duplicados. 
df_exp["client_id"].duplicated().sum()

#Vemos que no presenta duplicados, valor entre parentesis = 0
```

```python
#2.d Corregir el nombre de las columnas. 
df_exp = df_exp.rename(columns={"Variation": "variation"})
df_exp
```

```python
#2.e Convertir dataset limpio a CSV y guardar en carpeta Datasets_limpios
df_exp.to_csv("Datasets_limpios/clean_experiment_clients.csv", index=False)
```
